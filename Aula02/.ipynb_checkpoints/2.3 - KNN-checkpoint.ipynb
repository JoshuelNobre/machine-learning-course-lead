### KNN para classificação

```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
```

```python
knn = KNeighborsClassifier(n_neighbors = 3) # Cada elemento será comparado aos 3 vizinhos mais próximos dele
```

```python
data = pd.read_table("fruit_data_with_colors.txt")
```

```python
data
```

```python
X = data[['mass','width','height','color_score']]
y = data['fruit_label']
```

```python
X # Atributos
```

```python
y # Rótulos
```

#### Separando em treino e teste

```python
from sklearn.model_selection import train_test_split       #Impotando divisão de treino e teste
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y) # Por padrão 75% dos dados para treino e 25% para teste
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test) # Esse modelo acertou 66,6%
```

### Pré-procesando os dados para o KNN

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm = MinMaxScaler()
```

```python
X_train = mm.fit_transform(X_train) #Pegando os dados de treino e transformando suas colunas em dados entre 0 e 1
```

```python
X_train
```

```python
X_test = mm.transform(X_test)
```

```python
X_test
```

```python
knn = KNeighborsClassifier(n_neighbors=3) 
```

```python
knn.fit (X_train,y_train)
```

```python
knn.score(X_test,y_test) # Tivemos 86,6% de acerto
```

```python
knn.predict(X_test)
```
